```python
def alpha(df,multiplier,check_duration,hold_duration):
    net=0
    net_list=[]
    for i in range(check_duration+5,len(df['close'])-hold_duration-5):
        if df['close'][i]>np.mean(df['close'][i-check_duration:i-1])+multiplier*np.std(df['close'][i-check_duration:i-1]):
            net-=df['close'][i+hold_duration]-df['close'][i+1]
        if df['close'][i]<np.mean(df['close'][i-check_duration:i-1])-multiplier*np.std(df['close'][i-check_duration:i-1]):
            net+=df['close'][i+hold_duration]-df['close'][i+1]
        net_list.append(net)
    print(net)
    sharpe_ratio=sharpe(net_list)
    return [sharpe_ratio,net_list,net]
```

```python
def sharpe(net_list):
    temp=pd.DataFrame()
    temp['net']=net_list
    temp['pct']= temp['net'].pct_change(1)
    temp=temp.fillna(0)
    temp.replace([np.inf, -np.inf], 0, inplace=True)
    Sharpe_Ratio = temp['pct'].mean() / temp['pct'].std()
    return Sharpe_Ratio
```

```python
def objective(trial):
    m = trial.suggest_int("m", 3,6)
    c = trial.suggest_int("c", 5,30,5)
    h = trial.suggest_int("h", 2,30,5)
    sr=alpha(m,c,h)[2]
    return -sr
 
# We use the multivariate TPE sampler.
sampler = optuna.samplers.TPESampler(multivariate=True)
 
study = optuna.create_study(sampler=sampler)
study.optimize(objective, n_trials=50)
```
